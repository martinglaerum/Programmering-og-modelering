## Integrasjon og derivasjon

Analytisk derivasjon er det vi er vi bruker i matte. Når vi deriverer et utrykk vil vi alltid finne den nøyaktige verdien for den deriverte siden vi kan sette $\Delta x $ til å være uendelig liten. Datamaskiner kan ikke jobbe med uendelig små tall. Derfor må den bruke numerisk derivasjon. Numerisk derivasjon vil si at man finner 
Numerisk derivasjon er metoder som gir tilnærmede verdier av den deriverte. Ofte er det veldig stor nøyatighet når man bruker numerisk derivasjon, men det vil aldri være helt nøyaktig.

Et eksempel kan være $f(x) = x^2$.
Hvis vi vil finne $f'(2)$ analytisk får vi 4. Men hvis vi gjør det numerisk vil vi få verdier som 4.000023342 eller 3.999995466.



Numerisk integrasjon fungerer på samme måte som numerisk derivasjon. Vi bruker ulike metoder til å finne tilnærmede verdier av den integrerte.

Hvis vi bruker rektangelmetoden som eksempel kan vi se at hvis i gjør det analytisk kan vi sette $\Delta x$ til å være uendelig liten. Dette kan vi heller ikke gjøre nå. Hvis vi prøver å sette $\Delta x$ mindre enn $ 10^{-10} $ får vi avrundingsfeil og vi får veldig unøyaktige svar. Med en bredde på $ 10^{-10} $ vil vi få gode svar, men aldri helt riktige svar.

## $\Delta x$ og n

Når man dervierer står $\Delta x$ får avstanden mellom de to punktene man måler stignengen mellom. Ved analytisk derivasjon vil denne avstanden være 0. Ved bruk av numerisk derivasjon vil resultatet bli bedre jo mindre $\Delta x$ er, men hvis $\Delta x$ blir for liten, så blir det så mange desimaler at datamaskinen går i surr. Derfor får man aldri helt nøyaktige svar ved hjelp av en datamaskin.

I integrasjon vil n stå for antall rektangler man bruker. I analytisk Integrasjon går n mot uendelig. Det blir umulig for en datamaskin og sette et tall lik uendelig, derfor når vi bruker numerisk integrasjon så kan må vi sette n som et veldig stort tall. Hvis man setter n = 100 000 så holder ofte det til å få ganske nøyaktig svar.

## Oppgave 7.3

```python
from pylab import*

tid = []
antall = []
økning = []

def N(t):
    return 0.2*(e**t)+90

def deriv(f,x,dx=1E-8):
    return (f(x+dx)-f(x))/dx



for i in range(2,9):
    tid.append(i)
    antall.append(N(i))
    økning.append(deriv(N,i))
    
plot(tid,antall,"k")
plot(tid,økning,"red")
legend(["Antall planter", "endring av  antall planter"])
grid()
show()
```

## Oppgave 7.9

```python
from pylab import*

Tid = linspace(0,60,1000)
Temp = []
Endr = []

def T(t):
    return 70*(e**(-0.065*t))

def deriv(f,x,dx=1E-8):
    return (f(x+dx)-f(x))/dx

for i in range(1000):
    Temp.append(T(Tid[i]))
    Endr.append(deriv(T,(Tid[i])))

plot(Tid,Temp)
plot(Tid,Endr)
legend(["Temperatur","Endring av Temperatur"])
show()

print("Etter 43 timer synker temperaturen med",Endr[42])
```

```python
from pylab import*

data = loadtxt("posisjon.txt", skiprows = 2)

def deriv(x,t,i):
    dx = x[i]- x[i-1]
    dt = t[i] - t[i-1]
    return dx/dt

tid = data[:,0]

pos = data[:,1]

fart = []

aks = []

for i in range(999):
    fart.append(deriv(pos,tid,i))
    
for i in range(999):
    aks.append(deriv(fart,tid,i))

plot(tid,pos)
title("Posisjon")
show()
plot(tid[1:],fart[1:])
title("Fart")
show()
plot(tid[1:],aks[1:]) #hvis jeg satte tid[:2] ble grafen veldig rar. Ser uansett at aksjelerasjon er = 0.
title("Akselerasjon")
show()
```

## Oppgave 7.14

```python
from pylab import*


def f1(x):
    return 2*x+1

def f2(x):
    return (x**2)+4*x-5

def f3(x):
    return e**x

def f4(x):
    return (e**x)-5

def f5(x):
    return 6*(sqrt(x))-x

def f6(x):
    return 5*(e**(-2*x))

def f7(x):
    return sqrt(log(x))

def f8(x):
    return 4*log(x**2)

def f9(x):
    return 4*(5**((-x**3)-2*x))


def rekt(f, a=2, b=5, n=100):
    A1 = 0
    dx1 = (b-a)/n
    for k in range (1, n):
        A1 = A1 + f(a + (k*dx1))
    A1 = dx1*A1
    return A1


def trapes(f,N,A=2,B=5):
    A2 = 0
    dx2 = (B-A)/N
    for i in range(1,N):
        A2 = A2 + f(A+i*dx2)*dx2
    A2 = A2 + dx2*(f(A)+f(B))/2
    return A2


print(rekt(f1),trapes(f1,1000),trapes(f1,100000))
print(rekt(f2),trapes(f2,1000),trapes(f2,100000))
print(rekt(f3),trapes(f3,1000),trapes(f3,100000))
print(rekt(f4),trapes(f4,1000),trapes(f4,100000))
print(rekt(f5),trapes(f5,1000),trapes(f5,100000))
print(rekt(f6),trapes(f6,1000),trapes(f6,100000))
print(rekt(f7),trapes(f7,1000),trapes(f7,100000))
print(rekt(f8),trapes(f8,1000),trapes(f8,100000))
print(rekt(f9),trapes(f9,1000),trapes(f9,100000))
```

I kolonnen til venstre ble rektangelmetoden brukt og i kolonnen i midten ble trapesmetoden brukt. I kollonnen til høyre ble også trapesmetoden brukt, men her er antall trapeser mye større, det gjør at denne kolonnen inneholder en verdi som er nesten lik den analytiske verdien.

Vi kan se at ved bruk av samme antall rektangler/trapeser kommer trapesmetoden mye nærmere den analytiske verdien. Det viser at trapesmetoden er mye bedre og mye mer nøyaktig.
