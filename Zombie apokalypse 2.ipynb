# Introduksjon
Jeg valgte oppgave 1 der jeg skal programmere ulike modeller som kan beskrive hvordan en populasjon med mennekser klarer seg under en zombie-akopalypse. Framgangsmåten for oppgaven min er at jeg vil lage ulike utrykk som beskriver endring for et tilfelle f.eks. populasjon av mennesker. Videre vil jeg bruke eulers metode for å lage en graf som kan beskrive tilfellet.

## Teori
Eulers metode er en metode som tar utgangspunkt i endringen av noe f.eks. endringen til en befolkning, altså den deriverte av en fuksjon for befolkningen P'(t). Ved hjelp av et utrykk for endringen av befolkningen finner euler funksjonsverdiene for befolkningen P(t). Vi ser på en tilfeldig funksjon f(x). Ved å se på likningen for newtons kvotient kan vi se bekgrunnen for hvordan eulers metode fungerer. 

$$ f'(x) = \frac{f(x+h)-f(x)}{h}$$

Hvis vi gjør om på denne likningen kan vi får et utrykk for f(x+h).

$$ f(x+h) = f(x)+ f'(x) * h $$

Vanligvis når vi deriverer lar vi h gå mot null og får en nøyaktig verdi for den deriverte. Siden vi deriverer numerisk kan vi ikke få h til å gå mot null. Vi lar derimot h være et veldig lite tall helst mindre enn 10^-3. 


Vi ser at ved se på en verdi for den deriverte av en funksjon f'(t) og en verdi for samme funksjonen f(x) kan vi finne en ny verdi for f nemlig f(x+h). Siden vi har et et utrykk for f'(x) kan vi gjenta prossesen for f'(x+h) og f(x+h) for å finne f((x+h)+h) og slik kan vi forsette og finne verdier for f(x) for så å plotte funksjon. Det eneste vi trenger for å bruke denne metoden er den deriverte av en funksjon og startverdien for funksjonen f(0). 

I modellen jeg lager skal jeg modellere en populasjon. Derfor vil jeg beskrive endringen av en populasjon med en funksjon P'(t) og bruke dette til å plotte funksjonen for populasjonen, P(t). Jeg vil derfor forholde meg til dette utrykket.

$$ P(t+h) = P(t)+ P'(t) * h $$

Alle modellene i denne oppgaven er forenklede modeller av et tenkt scenario. Modellene er bare ment for å vise hvordan en populasjon ville klart seg under en zombieapokalypse og kan ikke brukes til å forutsi hvordan en populasjon ville klart seg i virkeligheten. En zombieapokalypse har aldri funnet sted så de ulike parameterne for zombiene i modellen er bare tenkt scenario der jeg ser på hva en zombie kan være og hvordan de kan oppføre seg hvis de faktisk var ekte.

## Modell 1
Under zombie-apokalypsen har 500 mennesker bosatt seg i landsbyen Alexandria. Vi ønsker å se på hvordan populasjonen til denne landsbyen utvikler seg de neste hundre årene. Først starter vi med en modell som viser hvordan populasjon vil øke hvis vi bare tar for oss fødselsrate og innvandring. 

Fødselsraten til befolkningen vil nok være ganske høy, da de er avhengig av en rask økning av populasjonen. Det er også sannsynlig at mange dør i ung alder grunnet dårlig tilgang til helsehjelp så det er viktig at de formerer seg mye. Den gjennomsnittlige fødselsraten til verden nå er ca. 2%. Siden disse menneksene prøver å ha en høy fødselsrate vil denne modellen  bruke en fødselsrate på 5% eller da 0.05. 

$$ Fødselsrate = 0.05 $$

Videre må vi se på innvandring til landsbyen. Det vil være mange mennesker som ikke har et sted å være under en akopalypse og mange av disse vil prøve å finne et trygt sted å bo. Derfor er det naturlig å tro at det vil være ganske høy innvandring. Det vil naturligvis ikke være en konstant mengde med innvandrere hvert år, men for å gjøre modellen litt enklere bruker jeg et konstant antall innvandrere. I denne modellen vil antall innvadrere vært år være 100.

$$ Innvandring = 100 $$

For at eulers metode skal fungere må vi også ha en startverdi for populasjonen. Siden populasjonen starter som 500 vil P(0) være  500.

$$ Populasjon = 500 $$

For å finne endringen i populasjonen må vi ta for oss disse to faktorene i et utrykk. Vi utrykker endringen til populasjonen som den deriverte av populasjonen, siden den deriverte betyr endringa av noe. 

$$ Pder = P*Fødselsrate + Innvandring $$

# Modell 5
Zombiene kan formere seg ved å drepe menneskene. Det vil si at de menneskene som ikke blir revet i fillebiter av zombiene blir til zombier. Hvis vi antar at 70% av menneskene som blir drept av zombiene blir til zombier kan vi lage et utrykk for endringen av zombiepopulasjonen. Siden vi økte dødsraten med 0.01 grunnet trusselen fra zombiene betyr det at 1% av populasjonen av mennesker dør fra zombieangrep. 

Siden vi satte den nye mengden innvandrere til halvparten av den tildligere mengden innvandrere betyr det at zombiene dreper den andre halvparten. Antall zombier som formerer seg fra angrep på innvandrerne vil derfor være den nye verdien for innvandring ganger 0.7 = 70%.

Vi bruker dette til å lage et utrykk for endring av zombiepopulasjon som er lik den deriverte av zombiepopulasjonen.

$$Zformeringsrate = 0.01*Populasjon*0.7 + Innvandring*0.7 $$

$$Zder = Zformeringsrate $$

# Modell 6
Siden zombiene kan drepe menneskene vil det være naturlig at menneskene prøver å forsvare seg. Når menneskene beveger seg utenfor landsbyen tar de meg seg våpen slik at de kan forsvare seg mot zombiene hvis zombiene prøver å anripe dem. Menneskene prøver også og lokke zombiene til veggene rundt landsbyen for så å drepe dem med spyd slik at de slipper å sette seg selv i fare. Vi må derfor innføre en slags dødsrate for zombiepopulasjonen. I og med at det er farlig å bevege seg nærme zombiene vil menneskene helst gjøre det i større grupper. Det betyr at jo flere mennesker som er i landsbyen jo flere slike grupper kan bli dannet. Vi legger derfor inn en dødsrate for zombiene som vil være avhengig av populasjonen av mennesker. I denne modellen bruker jeg at dødsraten for zombiene er 10% av menneskepopulasjonen.

$$Zdødsrate = 0.1*Populasjonen$$

$$Zder = Zformeringsrate - Zdødsrate  $$



# Modell 7
Alexandria er ikke den eneste levedyktige landsbyen som finnes i området. Andre grupper med mennesker har også greid bygge sine egen landsbyer ikke så langt unna Alexandria. Det finnes to slike landsbyer, Hilltop og Kingdom. Disse tre landsbyene kan hjelpe hverandre å drepe zombiene i området, og de kan også dele ressurser mellom seg. Det kan være at en av landsbyene har en type medisin som befolkningen i Alexandria trenger. Dette kan Alexandria bytte mot noe de andre landsbyene trenger. Det at det er tre landsbyer gjør det også lettere å bevege seg i området utenfor murene siden de kan danne større eskpedisjonsgrupper. Disse faktorene vil føre til at dødsraten til zombiene vil øke. Hvis vi antar at populasjonene er ca. like store kan vi tredoble dødsraten til zombiene.

$$Zdødsrate = Zdødsrate*3$$

Men siden det er tre ganger så mange mennesker kan vi også anta at formeringsraten til zombiene vil bli tredoblet. Siden de ulike landsbyene samarbeider kan det tenkes at menneskene vil bevege seg oftere utenfor murene, f.eks. hvis de skal dele ressurser med andre landsbyer, eller trenger hjelp. Vi antar derfor at antall ganger menneskene beveger seg utenfor murene til de ulike landsbyene vil dobles, noe som igjen fører til en dobling av zombieformeringsraten. 
$$ Zformeringsrate = Zformeringsrate * 3 * 2 $$

Siden menneskene beveger seg utenfor murene mer vil også dødsraten øke, ,men siden mennesken nå kan dele mediser og ressurser mellom seg fører dette til en reduksjon på dødsraten. Hvis risikoen for å øke antall turer utenfor landsbyen skal lønne seg må reduskjonen i dødsraten være større enn risikoen turene innebærer. Vi kan sette økningen av dødsraten fororsaket av den økte mengder turer til 0.66% men vi kan senke den med 1% takket være den økte tilgangen til ressurser.

$$Dødsrate = 0.03 + 0.0066 - 0.01$$

# Resultater

## Modell 1

```python
from pylab import*

# Variabler
Populasjon = 500
Fødselsrate = 0.05
Innvandring = 100

# Tidsparametere
Tid = 101 # Siden vi plotter T[:-1] gir dette at vi plotter 100 år.
h = 1E-4 # Bruker dx = h i denne løkken
N = N = int((100-0)/h) # antall intervaller

#Lister
P = zeros(N+1)
Pder = zeros(N+1)
T = zeros(N+1)

P[0] = Populasjon # Startverdien/Initialbetingelsen

for i in range(len(P)-1):  
    Pder[i] = P[i]*Fødselsrate + Innvandring
    P[i+1] = P[i] + Pder[i]*h
    T[i] = i*h

plot(T[:-1],P[:-1])
xlabel("Tid (År)")
ylabel("Antall mennesker")
legend(["Mennesker"])
show()
```

## Modell 2

```python
Bæreevne = 3000

for i in range(len(P)-1):  
    Pder[i] = (P[i]*Fødselsrate + Innvandring)* (1 - (P[i]/Bæreevne))
    P[i+1] = P[i] + Pder[i]*h
    T[i] = i*h

plot(T[:-1],P[:-1])
xlabel("Tid (År)")
ylabel("Antall mennesker")
legend(["Mennesker"])
show()
```

# Modell 3

```python
Dødsrate = 0.02

for i in range(len(P)-1):  
    Pder[i] = (P[i]*Fødselsrate + Innvandring - Dødsrate*P[i] ) * (1 - (P[i]/3000))
    P[i+1] = P[i] + Pder[i]*h
    T[i] = i*h
    
plot(T[:-1],P[:-1])
xlabel("Tid (År)")
ylabel("Antall mennesker")
legend(["Mennesker"])
show()
```

# Modell 4

```python
Z_Populasjon = 50000
Dødsrate = 0.03
Innvandrin = 50

#Lager lister for zombiene for å kunne plotte zombiepopulasjonen senere
Z = zeros(N+1)
Zder = zeros(N+1)
Z[0] = Z_Populasjon

for i in range(len(P)-1):  
    Pder[i] = (P[i]*Fødselsrate + Innvandring - Dødsrate*P[i]) * (1 - (P[i]/3000))
    P[i+1] = P[i] + Pder[i]*h
    T[i] = i*h
    Z[i] = Z[0] #Siden zombiepopulasjonen er konstant trenger vi ikke bruke eulers metode.

#Plotting av grafene    
fig = figure()
ax = fig.add_subplot(111)
data1 = ax.plot(T[:-1],P[:-1],"-b", label = "Mennesker")
ax2 = ax.twinx()
data2 = ax2.plot(T[:-1],Z[:-1],"-r", label =  "Zombier")

data = data1 + data2
datatittel = [l.get_label() for l in data]
ax.legend(data,datatittel, loc=0)

ax.grid()
ax.set_xlabel("Tid (År)")
ax.set_ylabel("Antell mennesker")
ax2.set_ylabel("Antall zombier")
ax2.set_ylim(0,60000)
ax.set_ylim(0,5000)
plt.show()
```

Siden grafen har en vekstfaktor uten å ha noen reduserende faktorerer får vi en eksposensiell graf. Denne modellen fungere derfor best for en populasjon som vokser eksposensiellt f.eks. en bakteriekoloni.
Siden populasjonen vi beskriver har manglende plass, en begrenset mengde ressurser og en dødsrate fungerer modellen dårlig i forhold til det vi prøver å modellere.

Alle variablene som er brukt i modellen kan også endres for å se på ulike utfall for zombieapokalypsen.

Hvis vi vil få en enda mer realistisk modell kan vi utvide modellen ved å se på noen faktorer som ikke er konstante men som skjer tilfeldig. Dette kan være en pandemi og sannsyligheten for at en pandemi skal oppstå. Det kan også være varierende værfohold. Varierende værfohold kan føre til at noen år er det dårlig jordbruksår som betyr at mengden mat minker og dødsraten vil øke takket være sult. Det kan også hende at det blir veldig kaldt et år som kan føre til mange av menneskene dør på grunn av at de ikke har god mulighet for å varme opp husene sine.
